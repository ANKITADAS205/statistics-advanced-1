1. A random variable in probability theory is a function that assigns a numerical value to each outcome in a sample space. It quantifies random phenomena and is categorized based on the type of values it can take.

2. There are two main types of random variables: discrete random variables (which take countable values like 0, 1, 2...) and continuous random variables (which take uncountable values within a given range, like any real number between 0 and 1).

3. Discrete distributions describe the probability of outcomes of a discrete random variable, assigning probabilities to specific values. Continuous distributions describe probabilities for continuous random variables, using probability density functions (PDFs) instead of assigning probabilities to specific values.

4. A probability distribution function (PDF) for a continuous random variable gives the likelihood of the variable falling within a particular range of values, and the total area under the curve equals 1. For discrete variables, it's often called the probability mass function (PMF).

5. The cumulative distribution function (CDF) gives the probability that a random variable is less than or equal to a certain value. Unlike the PDF, which gives the density at a point, the CDF accumulates probabilities up to that point.

6. A discrete uniform distribution assigns equal probabilities to all outcomes in a finite sample space. For example, rolling a fair die where each side (1 to 6) has a 1/6 chance.

7. The Bernoulli distribution models a random experiment with exactly two outcomes: success (usually coded as 1) and failure (coded as 0). Its key properties include a single trial, a probability $p$ of success, and an expected value of $p$.

8. The binomial distribution models the number of successes in a fixed number of independent Bernoulli trials. It's used when there are repeated experiments, each with the same probability of success.

9. The Poisson distribution models the number of times an event occurs in a fixed interval of time or space, when the events occur independently and at a constant average rate. It’s widely used in fields like traffic flow, call centers, and biology.

10. A continuous uniform distribution assigns equal probability density to all values within a specified interval. The probability is uniformly distributed across the range, like the outcome of selecting a number randomly between 0 and 1.

11. The normal distribution is a symmetric, bell-shaped curve centered around the mean. It is characterized by its mean (μ) and standard deviation (σ), and most data lies within three standard deviations from the mean.

12. The standard normal distribution is a special case of the normal distribution with a mean of 0 and a standard deviation of 1. It's important because many statistical tests and formulas are based on it, enabling easy comparison using Z-scores.

13. The Central Limit Theorem (CLT) states that the sampling distribution of the sample mean approaches a normal distribution as the sample size becomes large, regardless of the population's distribution. It underpins much of inferential statistics.

14. The Central Limit Theorem explains why normal distribution is so prevalent—it shows that averages of large samples from any distribution tend to be normally distributed, which allows for easier statistical inference.

15. Z-statistics are used in hypothesis testing to determine how far a sample mean deviates from the population mean in standard deviation units. It helps assess the significance of observed differences.

16. A Z-score is calculated as $Z = \frac{(X - \mu)}{\sigma}$, where $X$ is the value, $\mu$ is the mean, and $\sigma$ is the standard deviation. It shows how many standard deviations an element is from the mean.

17. A point estimate is a single value used to estimate a population parameter, like a sample mean. An interval estimate, like a confidence interval, provides a range that is likely to contain the parameter with a certain level of confidence.

18. Confidence intervals indicate the degree of uncertainty in a sample estimate. They provide a range within which the true population parameter is expected to lie with a specified probability, such as 95%.

19. A Z-score is used in confidence intervals to determine the margin of error around a point estimate. The Z-score for a given confidence level helps establish how far out from the mean the interval should extend.

20. Z-scores standardize values from different distributions, making them comparable. For example, test scores from different exams can be compared using their Z-scores to see which score is relatively higher.

21. Assumptions for applying the Central Limit Theorem include: a sufficiently large sample size (commonly n ≥ 30), independent observations, and finite variance in the population.

22. The expected value in a probability distribution is the theoretical average outcome, calculated as the weighted sum of all possible values multiplied by their probabilities. It reflects the long-run average if the experiment is repeated many times.

23. A probability distribution shows how probabilities are assigned to different values of a random variable. The expected outcome, or expected value, is derived from this distribution and indicates the central tendency of the random variable.


```python
#1
import numpy as np
random_var = np.random.rand()
print("Random Variable:", random_var)
```

```python
#2
import matplotlib.pyplot as plt

values = np.arange(1, 7)
pmf = [1/6] * 6

plt.bar(values, pmf)
plt.title('PMF of Discrete Uniform Distribution (Die Roll)')
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.grid(True)
plt.show()
```

```python
#3
from scipy.stats import bernoulli

def bernoulli_pdf(p, x):
    return bernoulli.pmf(x, p)

print("P(X=1):", bernoulli_pdf(0.6, 1))
print("P(X=0):", bernoulli_pdf(0.6, 0))
```

```python
#4
binom_data = np.random.binomial(n=10, p=0.5, size=1000)

plt.hist(binom_data, bins=np.arange(12)-0.5, edgecolor='black', density=True)
plt.title('Binomial Distribution (n=10, p=0.5)')
plt.xlabel('Successes')
plt.ylabel('Probability')
plt.grid(True)
plt.show()
```

```python
#5
from scipy.stats import poisson

x = np.arange(0, 15)
pmf = poisson.pmf(x, mu=4)

plt.bar(x, pmf)
plt.title('Poisson Distribution (λ = 4)')
plt.xlabel('k')
plt.ylabel('P(X=k)')
plt.grid(True)
plt.show()
```

```python
#6
uniform_pmf = [1/6] * 6
cdf_uniform = np.cumsum(uniform_pmf)

plt.step(values, cdf_uniform, where='mid')
plt.title('CDF of Discrete Uniform Distribution')
plt.xlabel('Outcome')
plt.ylabel('Cumulative Probability')
plt.grid(True)
plt.show()
```

```python
#7
data = np.random.uniform(low=0, high=1, size=1000)
plt.hist(data, bins=20, edgecolor='black', density=True)
plt.title('Continuous Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

```python
#8
normal_data = np.random.normal(loc=0, scale=1, size=1000)
plt.hist(normal_data, bins=30, edgecolor='black', density=True)
plt.title('Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

```python
#9
from scipy.stats import zscore

z_scores = zscore(normal_data)
plt.hist(z_scores, bins=30, edgecolor='black', density=True)
plt.title('Z-Scores Distribution')
plt.xlabel('Z-Score')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

```python
#10
samples = [np.mean(np.random.exponential(scale=2, size=30)) for _ in range(1000)]
plt.hist(samples, bins=30, edgecolor='black', density=True)
plt.title('CLT on Exponential Distribution')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

```python
#15
means = [np.mean(np.random.normal(0, 1, 30)) for _ in range(1000)]
plt.hist(means, bins=30, edgecolor='black', density=True)
plt.title('CLT Verification')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

```python
#16
from scipy.stats import norm

x = np.linspace(-4, 4, 1000)
pdf = norm.pdf(x)

plt.plot(x, pdf)
plt.title('Standard Normal Distribution (μ=0, σ=1)')
plt.xlabel('Z')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

```python
#17
probabilities = [np.random.binomial(n=10, p=0.5) for _ in range(10)]
print("Binomial Outcomes:", probabilities)
```

```python
#18
data_point = 75
mean = 70
std_dev = 10

z = (data_point - mean) / std_dev
print("Z-score:", z)
```

```python
#19
sample_mean = 52
pop_mean = 50
std_error = 2
z = (sample_mean - pop_mean) / std_error

print("Z-statistic:", z)
```

```python
#20
import scipy.stats as stats

sample = np.random.normal(loc=50, scale=5, size=100)
mean = np.mean(sample)
std_err = stats.sem(sample)
conf_int = stats.t.interval(0.95, df=len(sample)-1, loc=mean, scale=std_err)

print("95% Confidence Interval:", conf_int)
```

```python
#21
data = np.random.normal(60, 8, 100)
mean = np.mean(data)
se = stats.sem(data)
ci = stats.t.interval(0.95, len(data)-1, loc=mean, scale=se)
print("Confidence Interval for Mean:", ci)
```
